## **Membuat Statistik Deskriptif**

```python
# Import Library yang Dibutuhkan
import numpy as np
import matplotlib.pyplot as plt
```

### **Kolom Positif**

**Rata-Rata (Mean)**
Mean adalah rata-rata matematika sederhana dari sekumpulan dua atau lebih bilangan. Mean menunjukkan distribusi nilai yang sama untuk kumpulan data tertentu. Untuk menghitung mean, kamu perlu menambahkan nilai total yang diberikan dalam data dan kemudian membagi jumlahnya dengan jumlah nilai total. 
Rumus mean adalah

*mean = jumlah data : banyak data*

Contoh:
4, 5, 8, 9, 12


mean=(4+5+8+9+12):5= 7.6

Rumus Matematis nya sebagai berikut.


**x̄ = ∑x / n**


Keterangan :


x̄ = Rata-rata (mean)

∑x = Jumlah total data

n = Banyaknya data

```python
# mencari rata-rata
c.Positif.mean()
```

```python
# mencari nilai tengah
c.Positif.median()
```

```python
# mencari data yang paling banyak muncul
c.Positif.mode()
```

```python
# mencari variance
c.Positif.var()
```

```python
# mencari standar deviasi
c.Positif.std()
```

```python
# mencari rentang data
c.Positif.max() - c.Positif.min()
```

```python
# melihat sebaran menggunakan boxplot
c.boxplot(column='Positif', return_type='axes')
```

## **Membuat Skewned Data**

```python
#import library yang dibutuhkan
import seaborn as sns
```

### **Kolom Positif**

```python
dt_Positif = c['Positif']
dt_Positif.skew()
```

```python
plt.figure()
sns.distplot(dt_Positif)
plt.show()
```

### **Kolom Odp**
